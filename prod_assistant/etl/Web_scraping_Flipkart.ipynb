```python
# 1. Get your lists using the classes you confirmed are working
names = soup.find_all('div', class_='RG5Slk')
prices = soup.find_all('div', class_='hZ3P6w DeU9vF')
ratings = soup.find_all('div', class_='MKiFS6')

itemlist = []

# 2. Use zip to iterate through all three lists at the same time
# zip() stops at the shortest list, ensuring no "NoneType" errors
for name, price, rate in zip(names, prices, ratings):
    item_dict = {
        'name': name.text.strip(),
        'price': price.text.strip(),
        'rate': rate.text.strip()
    }
    itemlist.append(item_dict)
    print(f"Scraped: {item_dict['name']} | {item_dict['price']} | {item_dict['rate']}")

print(f"\n--- Total items combined: {len(itemlist)} ---")
```

```python
import csv
import time
import re
import os
import pandas as pd
from bs4 import BeautifulSoup
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

class FlipkartScraper:
    def __init__(self, output_dir="data"):
        self.output_dir = output_dir
        if not os.path.exists(self.output_dir):
            os.makedirs(self.output_dir, exist_ok=True)

    def get_top_reviews(self, product_url, count=2):
        """Visits product page to extract deep review text from G4PxIA."""
        options = uc.ChromeOptions()
        options.add_argument("--headless")
        driver = uc.Chrome(options=options, use_subprocess=True)

        if not product_url.startswith("http"):
            driver.quit()
            return "No reviews found"

        reviews = []
        try:
            driver.get(product_url)
            time.sleep(3)
            driver.execute_script("window.scrollBy(0, 1000);")
            time.sleep(1)

            soup = BeautifulSoup(driver.page_source, "html.parser")
            # Using your G4PxIA selector for reviews
            review_blocks = soup.select("div.G4PxIA")
            
            for block in review_blocks:
                text = block.get_text(strip=True).replace("READ MORE", "")
                if len(text) > 10:
                    reviews.append(text)
                if len(reviews) >= count:
                    break
        finally:
            driver.quit()
        return " || ".join(reviews) if reviews else "No reviews found"
    
    def scrape_flipkart_products(self, query, max_products=2):
        options = uc.ChromeOptions()
        driver = uc.Chrome(options=options, use_subprocess=True)
        
        search_url = f"https://www.flipkart.com/search?q={query.replace(' ', '+')}"
        driver.get(search_url)
        
        try:
            # 1. Wait for your verified Item container: jIjQ8S
            wait = WebDriverWait(driver, 15)
            wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.jIjQ8S")))
            
            # 2. Extract containers
            items = driver.find_elements(By.CSS_SELECTOR, "div.jIjQ8S")[:max_products]
            print(f" Found {len(items)} product containers using jIjQ8S.")

            results = []
            for item in items:
                try:
                    # Verified Selectors from your session
                    name = item.find_element(By.CSS_SELECTOR, "div.RG5Slk").text.strip()
                    price = item.find_element(By.CSS_SELECTOR, "div.hZ3P6w.DeU9vF").text.strip()
                    rating = item.find_element(By.CSS_SELECTOR, "div.MKiFS6").text.strip()
                    
                    # Span selectors for totals
                    total_ratings = item.find_element(By.CSS_SELECTOR, "span.PvbNMB").text.strip()
                    total_reviews = item.find_element(By.CSS_SELECTOR, "span.o2SIOJ").text.strip()
                    
                    link = item.find_element(By.CSS_SELECTOR, "a").get_attribute("href")
                    
                    print(f" Fetching deep reviews for: {name[:30]}...")
                    top_reviews = self.get_top_reviews(link)

                    results.append({
                        "Model": name,
                        "Price": price,
                        "Rating": rating,
                        "Total Ratings": total_ratings,
                        "Total Reviews": total_reviews,
                        "Top Reviews": top_reviews,
                        "URL": link
                    })

                except Exception as e:
                    print(f"    Skipping item: {e}")
                    continue

            return pd.DataFrame(results)

        finally:
            driver.quit()
            print(" Browser closed.")

# --- RUN THE IPYNB CELL ---
scraper = FlipkartScraper()
df = scraper.scrape_flipkart_products("iphone 16", max_products=2)

if not df.empty:
    display(df) # Renders clean table in Notebook
    df.to_csv("data/flipkart_final_data.csv", index=False)
    print(" Successfully saved to data/flipkart_final_data.csv")
```

```python
import csv
import time
import re
import os
import pandas as pd
from bs4 import BeautifulSoup
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

class FlipkartScraper:
    def __init__(self, output_dir="data"):
        self.output_dir = output_dir
        if not os.path.exists(self.output_dir):
            os.makedirs(self.output_dir, exist_ok=True)

    def get_top_reviews(self, product_url, count=3):
        options = uc.ChromeOptions()
        options.add_argument("--headless")
        options.add_argument("--no-sandbox")
        driver = uc.Chrome(options=options, use_subprocess=True)

        reviews = []
        try:
            driver.get(product_url)
            time.sleep(3)
            driver.execute_script("window.scrollBy(0, 1000);")
            time.sleep(1)

            soup = BeautifulSoup(driver.page_source, "html.parser")
            # Using your G4PxIA review class
            review_blocks = soup.select("div.G4PxIA")
            for block in review_blocks:
                text = block.get_text(separator=" ", strip=True)
                if len(text) > 20:
                    reviews.append(text)
                if len(reviews) >= count:
                    break
        finally:
            driver.quit()
        return " || ".join(reviews) if reviews else "No reviews found"
    
    def scrape_flipkart_products(self, query, max_products=4):
        options = uc.ChromeOptions()
        driver = uc.Chrome(options=options, use_subprocess=True)
        
        search_url = f"https://www.flipkart.com/search?q={query.replace(' ', '+')}"
        driver.get(search_url)
        
        try:
            wait = WebDriverWait(driver, 15)
            # Wait for your verified item container jIjQ8S
            wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.jIjQ8S")))
            
            items = driver.find_elements(By.CSS_SELECTOR, "div.jIjQ8S")[:max_products]
            data_rows = []

            for item in items:
                try:
                    # 1. Product ID
                    pid = item.get_attribute("data-id")
                    
                    # 2. Product Title (RG5Slk)
                    title = item.find_element(By.CSS_SELECTOR, "div.RG5Slk").text.strip()
                    
                    # 3. Rating (MKiFS6)
                    rating = item.find_element(By.CSS_SELECTOR, "div.MKiFS6").text.strip()
                    
                    # 4. Total Reviews (o2SIOJ) -> STRICT NUMERIC LOGIC
                    try:
                        raw_rev = item.find_element(By.CSS_SELECTOR, "span.o2SIOJ").text.strip()
                        # This line removes commas and extracts ONLY digits
                        # Example: "(4,695 Reviews)" becomes "4695"
                        numeric_string = "".join(re.findall(r'\d+', raw_rev.replace(',', '')))
                        total_reviews = int(numeric_string) if numeric_string else 0
                    except:
                        total_reviews = 0
                        
                    # 5. Price (hZ3P6w.DeU9vF)
                    price = item.find_element(By.CSS_SELECTOR, "div.hZ3P6w.DeU9vF").text.strip()
                    
                    # 6. Top Reviews
                    link = item.find_element(By.CSS_SELECTOR, "a").get_attribute("href")
                    if not pid:
                        pid = re.search(r"/p/([^/?]+)", link).group(1) if "/p/" in link else "N/A"
                        
                    print(f"Scraping: {title[:40]}... [Reviews: {total_reviews}]")
                    top_reviews = self.get_top_reviews(link)

                    data_rows.append([pid, title, rating, total_reviews, price, top_reviews])

                except Exception as e:
                    print(f" Error: {e}")
                    continue

            return data_rows
        finally:
            driver.quit()

    def save_to_csv(self, data, filename="product_reviews.csv"):
        path = os.path.join(self.output_dir, os.path.basename(filename))
        with open(path, "w", newline="", encoding="utf-8") as f:
            writer = csv.writer(f)
            writer.writerow(["product_id", "product_title", "rating", "total_reviews", "price", "top_reviews"])
            writer.writerows(data)
        print(f" CSV saved to: {path}")

# --- EXECUTION ---
scraper = FlipkartScraper()
final_data = scraper.scrape_flipkart_products("iphone 16 pro max", max_products=4)

if final_data:
    scraper.save_to_csv(final_data, "flipkart_iphone_final.csv")
    # Preview as DataFrame
    df = pd.DataFrame(final_data, columns=["product_id", "product_title", "rating", "total_reviews", "price", "top_reviews"])
    display(df)
```
